```python
# This allows us to use matplotlib inline
%matplotlib inline
from matplotlib.pyplot import *

import numpy as np
```

$$
\sum_{i=1}^{360} x 1.08^{360-i}
$$

```python
def calc_equity(months, interest, savings):
    return [np.sum((savings) * ((1+interest)**(1/12.0))**(m-np.arange(0, m, 1))) for m in np.arange(0, months, 1)]
```

```python
equity = calc_equity(months=360, interest=0.08, savings=100)
equity_broker = calc_equity(months=360, interest=0.08, savings=100-20)
equity_fund = calc_equity(months=360, interest=0.06, savings=100)
```

```python
figure(figsize=(5, 5))

plot(equity, 'b')
plot(equity_broker, 'g')
plot(equity_fund, 'r')

xticks([0, 12*5, 12*10, 12*15, 12*20, 12*25, 12*30])
gca().set_xticklabels([0, 5, 10, 15, 20, 25, 30]);

savefig('fig.svg')
```

```python
print("equity {:,.2f} \n equity_broker {:,.2f} \n equity_fund {:,.2f}".format(
    equity[-1],
    equity_broker[-1],
    equity_fund[-1],
))
```
